<h1>Model</h1>

```python
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
clf_names = [
     "Nearest Neighbors", 
     "Decision Tree", 
     "Random Forest", 
     "AdaBoost",
     "Naive Bayes", 
     "Neural Net",
     "Linear SVM", 
     "RBF SVM",
]

classifiers = [
    KNeighborsClassifier(5),
    DecisionTreeClassifier(max_depth=20),
    RandomForestClassifier(max_depth=20, n_estimators=10, max_features=2),   
    AdaBoostClassifier(),
    GaussianNB(),
    MLPClassifier(alpha=1, max_iter=1000),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1)
]
```

```python
import time

x_train, x_val, y_train, y_val = \
    train_test_split(train_data_topk, train_data_label, test_size=.15, random_state=42)

for name, clf in zip(clf_names, classifiers):
    start_time = time.time()
    clf.fit(x_train, y_train)
    print("Classifer {} Training time: {:.2f} second".format(name, time.time() - start_time))
```

<h1>Evaluation</h1>

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import itertools
```

```python
np.seterr(divide='ignore',invalid='ignore')
labels = ["Normal", "Probe", "DOS", "U2R", "R2L"]
```

```python
def plot_confusion_matrix(cm, target_names, title, normalize=True, cmap=plt.cm.Blues):
    accuracy = np.trace(cm) / float(np.sum(cm))
    
    fig, ax = plt.subplots()
    
    n_classes = cm.shape[0]

    im_ = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    cmap_min, cmap_max = im_.cmap(0), im_.cmap(256)
    plt.title("{}: {:.2%}".format(title, accuracy), fontsize=14)
          
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    thresh = 0.99 if normalize else cm.max() / 2
    for i, j in itertools.product(range(n_classes), range(n_classes)):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     ha="center", va="center",
                     color=cmap_min if cm[i, j] > thresh else cmap_max)
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     ha="center", va="center",
                     color=cmap_min if cm[i, j] > thresh else cmap_max)
            
    fig.colorbar(im_, ax=ax)

    ax.set(xticks=np.arange(n_classes),
       yticks=np.arange(n_classes),
       xticklabels=target_names,
       yticklabels=target_names,
       ylabel="True label",
       xlabel="Predicted label")
    ax.set_ylim((n_classes - 0.5, -0.5))
    plt.setp(ax.get_xticklabels(), rotation='horizontal')

    plt.show()
```

```python
for name, clf in zip(clf_names, classifiers):
    start_time = time.time()
    y_pred = clf.predict(x_val)
    
    cm = confusion_matrix(y_val, y_pred, labels=labels)
    plot_confusion_matrix(cm, labels, name, False)
```
